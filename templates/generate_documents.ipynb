```python
localhost_directory = "/Library/WebServer/Documents/"
root_directory = "../"
web_directory = "https://nkhklkstksch.xyz/"
# root_directory = "/Users/nkhklkstksch/Desktop/store_active/repositories/public_portfolio.github.io/"
document_layout = "layout"
document_style = "style"

document_layout = open(f"{root_directory}templates/{document_layout}.xml", "r").read()

handle_style = open(f"{root_directory}templates/{document_style}.css", "r").read()
root_style = open(f"{root_directory}templates/{document_style}.css", "w")
root_style.write(handle_style)
root_style.close()
localhost_style = open(f"{localhost_directory}templates/{document_style}.css", "w")
localhost_style.write(handle_style)
localhost_style.close()
```

```python
content_title = "index"
content_language = "en"
content_body = open(f"{root_directory}index.xml", "r").read()
content_type = "board"

root_document = document_layout.format(
    language=content_language,
    directory=web_directory,
    title=content_title,
    type=content_type,
    body=content_body,)
write_root = open(f"{root_directory}index.html", "w")
write_root.write(root_document)
write_root.close()

localhost_document = document_layout.format(
    language=content_language, 
    directory="./",
    title=content_title,
    type=content_type,
    body=content_body,)
write_localhost = open(f"{localhost_directory}index.html", "w")
write_localhost.write(localhost_document)
write_localhost.close()
```
